```python
import pandas as pd

data_train=pd.read_csv("../data/train.csv")
data_train.info()
```

```python
data_train.isnull().sum()
```

```python
to_int=["Delivery_person_Ratings","Delivery_person_Age"]
for i in to_int:
    data_train[i] = pd.to_numeric(data_train[i], errors='coerce')
    median_val = data_train[i].median()
    data_train[i] = data_train[i].fillna(median_val).astype(int)    
    
```

```python
data_train['Delivery_person_Ratings'].value_counts()
```

```python
to_float=["Delivery_location_longitude","Delivery_location_latitude","Restaurant_longitude","Restaurant_latitude"]
for f in to_float:
    data_train[i] = data_train[i].fillna(0).astype(float)

```

```python
data_train['Delivery_location_longitude'].value_counts()
```

```python
data_train['Time_Orderd']        = pd.to_datetime(data_train['Time_Orderd'],        format='%H:%M:%S', errors='coerce')
data_train['Time_Order_picked']  = pd.to_datetime(data_train['Time_Order_picked'],  format='%H:%M:%S', errors='coerce')

# Compute prep duration in minutes
 
data_train['Preparation_Duration'] = (data_train['Time_Order_picked'] - data_train['Time_Orderd']).dt.total_seconds() / 60

# Handle negative values (midnight crossing e.g. 23:50 -> 00:05)
data_train.loc[data_train['Preparation_Duration'] < 0, 'Preparation_Duration'] += 1440
```

```python
data_train['Preparation_Duration'].value_counts()
```

```python
data_train["City"].value_counts()
```

```python
data_train["multiple_deliveries"] = pd.to_numeric(data_train["multiple_deliveries"], errors='coerce')
data_train["multiple_deliveries"] = data_train["multiple_deliveries"].fillna(4).astype(int)
```

```python
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder

# ohe_categories=["Weatherconditions","Type_of_vehicle"]
ctgr_str=["Festival","Type_of_vehicle","Type_of_order","Weatherconditions"]
ohe = OneHotEncoder(sparse_output=False)
encoded=ohe.fit_transform(data_train[ctgr_str])
encoded_df=pd.DataFrame(encoded,columns=ohe.get_feature_names_out(ctgr_str))
data_train = data_train.drop(columns=ctgr_str)
data_train = pd.concat([data_train, encoded_df], axis=1) 


oe=OrdinalEncoder()
data_train["Road_traffic_density"] = oe.fit_transform(
    data_train[["Road_traffic_density"]]
)
```

```python
# 1. Convert to string and strip hidden spaces
data_train["Festival"] = data_train["Festival"].astype(str).str.strip()

# 2. Perform the replacement and save it
data_train["Festival"] = data_train["Festival"].replace({"NaN": "in progress"})

# 3. Verify
print(data_train["Festival"].value_counts())
```

```python
data_train["Time_taken(min)"] = data_train["Time_taken(min)"].str.replace("(min) ", "", regex=False)
data_train["Time_taken(min)"].astype(int)
```

```python
data_train["Weatherconditions"] = data_train["Weatherconditions"].astype(str).str.strip()

data_train["Weatherconditions"] = data_train["Weatherconditions"].replace({"conditions NaN":"conditions Moderate"})

print(data_train["Weatherconditions"].value_counts())
```

```python

oe=OrdinalEncoder()
data_train["Road_traffic_density"] = oe.fit_transform(
    data_train[["Road_traffic_density"]]
)
```

```python
data_train["Time_taken(min)"]=data_train["Time_taken(min)"].astype(int)
```

```python
to_drop=["Time_Order_picked","Time_Orderd","Order_Date","ID","Order_Date","City"]
data_train = data_train.drop(columns=to_drop)
```
